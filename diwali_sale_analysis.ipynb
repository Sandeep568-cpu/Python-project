```python
def categorise_age(age):
    if age <= 17:
        return '0-17'
    elif age<= 25:
          return '18-25'
    elif age<= 35:
          return '26-35'
    elif age<=45:
          return '35-45'
    elif age<= 50:
          return '45-50'
    elif age<= 55:
          return '45-55'
    else:
        return '55+'
```

```python
df['Age Group']=df['Age'].apply(categorise_age)
```
